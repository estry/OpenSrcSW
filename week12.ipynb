```python
# 이용객이 가장 많은 역 순위
# 1. 역명(이용객 수)
# 2. 역명(이용객 수)
# 3. 역명(이용객 수)
# 이용객이 가장 적은 역 순위
# 1. 역명(이용객 수)
# 2. 역명(이용객 수)
# 3. 역명(이용객 수)
import pandas as pd

df = pd.read_csv('subwaytraffic.csv', encoding='cp949')
line_list = df['호선명'].unique()
df['승차승객수'] = df['승차승객수'].apply(lambda x : x.replace(',','_'))
df['하차승객수'] = df['하차승객수'].apply(lambda x : x.replace(',','_'))
df['승차승객수'] = df['승차승객수'].astype('int')
df['하차승객수'] = df['하차승객수'].astype('int')
df['총이용객'] = df['승차승객수'] + df['하차승객수']
print('이용객이 가장 많은 역 순위')
for line in line_list:
    temp_df = df[df['호선명'] == line].sort_values(by='총이용객', ascending=False)
    temp_df = temp_df.reset_index()
    print(temp_df[['호선명','지하철역','총이용객']].head(3))
print()
print('이용객이 가장 적은 역 순위')
for line in line_list:
    temp_df = df[df['호선명'] == line].sort_values(by='총이용객', ascending=True)
    temp_df = temp_df.reset_index()
    print(temp_df[['호선명','지하철역','총이용객']].head(3))
```
